### Análise descritiva da base

```python
#import das bibliotecas usadas

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/Time Americans Spend Sleeping.csv")
```

```python
df.head()
```

### Sobre o dataset

O conjunto de dados fornece informações sobre a quantidade média de tempo que os americanos passam dormindo.

- **Year**: O ano em que os dados foram coletados. 

- **Period**: o período do dia em que os dados foram coletados.

- **Avg hrs per day sleeping**: O número médio de horas por dia que os americanos passam dormindo. 

- **Standard Error**: O erro padrão para o número médio de horas por dia que os americanos passam dormindo. 

- **Type of Days**: o tipo de dia em que os dados foram coletados. 

- **Age Group**: a faixa etária dos americanos pesquisados. 

- **Activity**: a atividade em que os americanos pesquisados ​​estavam engajados. 

- **Sex**: O sexo dos americanos pesquisados.

## Visualização em pandas

```python
#df_novo=df.pivot_table(index="Year",values="Avg hrs per day sleeping",aggfunc="mean",sort=False).plot(kind = "bar")
ax = df.plot(y = "Avg hrs per day sleeping", kind = "hist", legend = None, figsize = (9, 7), bins=12, alpha=0.6, grid = 1, edgecolor = "black")
ax.set_title("Relação das horas dormidas", fontsize = 18)
ax.set_xlim(7.5, 10.8)
ax.set_ylabel(None)
ax = ax
```

## Visualização em matplotlib

```python
#visualização dos anos 
lista_anos = df["Year"].unique()
lista_anos
```

```python
#visualização das faixas etárias
lista_age_group = df["Age Group"].unique()
print(lista_age_group)

#quantidade de pessoas em cada faixa etária
qtd_age_group = df["Age Group"].value_counts()
print(qtd_age_group)
```

```python
plt.figure(figsize=(14, 9), facecolor="#F5FFFA")
plt.pie(qtd_age_group, data = df, 
        autopct = "%.1f%%", 
        colors=['#7FFFD4', '#66CDAA', '#5F9EA0', '#00CED1', '#40E0D0', '#48D1CC', '#20B2AA'],
        wedgeprops = {"edgecolor" : "gray", 
                      'linewidth': 1, 
                      'antialiased': True}) 
plt.legend(labels = lista_age_group, loc='upper left', bbox_to_anchor=(0.8, 0.7))
plt.title("Entrevistados por faixa etária", fontsize = 18, y = 0.95, x = 0.55)
plt.show()
```

```python
df_novo=df.pivot_table(index=colunas_int[0],values=colunas_float[0],aggfunc="mean",sort=False)
```

```python
plt.figure(figsize=(12,7))
ax=sns.lineplot(df_novo.index,df_novo[colunas_float[0]],color="black")
#ax.set_title("Colocar o titulo", fontsize=20)
ax.set_xlabel("Anos", fontsize=15)
ax.set_ylabel("Media de horas de sono", fontsize=15)
plt.show()
```

### Visualização em seaborn

```python
ax = sns.catplot(x = "Type of Days", y = "Avg hrs per day sleeping", data = df, height=6, aspect=1.5)
plt.title("Horas dormidas por dia x Tipo de dia", fontsize = 18)
ax = ax
```

```python
plt.figure(figsize=(12,7))
plt.hist(bins=10,x=df[colunas_float[0]],alpha = 0.5,color="red", edgecolor = "black",)
plt.xlabel("Horas de sono", fontsize=15)
plt.show()
```

### Visualização em plotly

```python
fig=px.histogram(df['Avg hrs per day sleeping'],facet_col=df["Sex"],facet_row=df["Type of Days"],color=df["Sex"])
#titulos no eixo x
fig.update_xaxes(title_text="Horas de sono", row=1, col=1)
fig.update_xaxes(title_text="Horas de sono", row=1, col=2)
fig.update_xaxes(title_text="Horas de sono",  row=1, col=3)

#titulos no eixo y
fig.update_yaxes(title_text="frequência", row=1, col=1)
fig.update_yaxes(title_text="frequência", row=2, col=1)
fig.update_yaxes(title_text="frequência", row=3, col=1) 

fig.update_layout(
     
    legend_title={
        'text':'Variavel',
        'font_size': 16
    },
    font_color='black',
    title={
        'text': 'Frequencia de horas dormidas',
        'x': 0.30,
        'y': 0.93,
        'font_family': 'Arial',
        'font_color': 'black',
        'font_size': 20
    },
)
fig.update_layout(annotations=[{
    'text':''
}])
fig
```

```python
df1 = df.pivot_table(index = "Type of Days", columns = "Sex", values = "Avg hrs per day sleeping")

ax = px.bar(df1)

ax.update_layout(
    xaxis_title={
        'text': None
    },
    yaxis_title={
        'text': None
    },
    legend_title={
        'text':'Sexo',
        'font_size': 16
    },
    font_color='gray',
    title={
        'text': 'Relação das horas dormidas por Tipo de dia e Sexo',
        'x': 0.5,
        'y': 0.93,
        'font_family': 'Arial',
        'font_color': 'gray',
        'font_size': 20
    }
)
```
